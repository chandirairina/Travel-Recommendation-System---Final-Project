# Menyesuaikan simbol-simbol agar tidak mengganggu dalam proses cleaning data

```python
df["Things to Do"]=df["Things to Do"].str.replace("[","").str.replace("]","").str.replace("'","").str.replace(" ","").str.replace("(",",(").str.replace(",Lihatsemua,",",").str.replace(",Lihatsemua","").str.replace('&',"dan").str.replace("/","atau").str.replace(".","").str.replace("-","").str.replace("'","").str.replace('"',"")
```

```python
df
```

```python
df["Things to Do"]=df["Things to Do"].str.split(",")
```

```python
df
```

```python
todo=[]

for j in range(len(df)):
    inside_todo=[]
    
    for i in df["Things to Do"][j]:
        if "(" in i:
            inside_todo.append(" ")
        else:
            inside_todo.append(i)
    todo.append(inside_todo)
```

```python
len(todo)
```

```python
df["Things to Do"]=todo
```

```python
df
```

```python
for i in range(len(df)):
    df["Things to Do"][i]=''.join([str(elem) for elem in df["Things to Do"][i]])
```

```python
df
```

```python
df["Things to Do"]=df["Things to Do"].str.lower()
```

```python
df.to_csv("dataset_others_engineered.csv")
```
